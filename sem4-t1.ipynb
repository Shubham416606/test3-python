```python
import networkx as nx
import matplotlib.pyplot as graph
G=nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,6),(4,6),(5,2)])
nx.draw_networkx(G,node_size=850,node_color='red',width=5,edge_color='blue')
graph.show()
```

```python
import networkx as nx
import matplotlib.pyplot as graph
G=nx.DiGraph()
l=[[1,0,1],[1,1,0],[0,1,1]]
G.add_nodes_from([0,1,2])
for i in range(0,3):
    for j in range(0,3):
        if l[i][j]==1:
            G.add_edge(i,j)
lable={0:'emp1',1:'emp2',2:'emp3'}
nx.draw_networkx(G,node_size=850,node_color='red',width=5,edge_color='blue',labels=lable,with_labels=True)
graph.show()
```

```python
import numpy as np
months =np.arange(12)
income=np.array([5,9,6,6,10,7,6,4,5,6,4,5])
expense=np.array([6,6,8,3,6,9,7,8,6,6,4,8])
fig,ax=plt.subplots(figsize=(8,8))
ax.plot(months,income,color='green')
ax.plot(months,expense,color='red')
ax.fill_between(months,income,expense,where=(income>expense),color='green',label='positive',alpha=0.3,interpolate=True)
ax.fill_between(months,income,expense,where=(income<expense),color='red',label='nagetive',alpha=0.3,interpolate=True)
plt.xlabel('months')
plt.ylabel('income')
plt.legend()
plt.show()
```
